```python
list=list.values.tolist()
```

```python
for i in range(0,len(list)):
    list[i]=[item for item in list[i] if not(pd.isnull(item)) == True]
    i=i+1
```

## Preprocessing

```python
encoder1=TransactionEncoder()
encoder1_arr=encoder1.fit(list).transform(list)
df1=pd.DataFrame(encoder1_arr,columns=encoder1.columns_)
df1
```

```python
df1.sum(axis=0)
```

## Model

### Apriori

```python
frequent_itemsets = apriori(df1, min_support=0.10, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
frequent_itemsets.dtypes
```

```python
frequent_itemsets[ (frequent_itemsets['length'] >= 2) &
                   (frequent_itemsets['support'] >= 0.10) ]
```

```python
for i in range(0,len(frequent_itemsets)):
    for s in frequent_itemsets['itemsets'][i]:
        if s=='5576':
            print(frequent_itemsets[frequent_itemsets.index==i])
    i=i+1
```

### FP Growth

```python
frequent_itemsets1 = fpgrowth(df1, min_support=0.1, use_colnames=True)
frequent_itemsets1['length'] = frequent_itemsets1['itemsets'].apply(lambda x: len(x))
frequent_itemsets1
```

```python
frequent_itemsets1[ (frequent_itemsets1['length'] >= 2) &
                   (frequent_itemsets1['support'] >= 0.10) ]
```

## Evaluation

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=5)
rules[(rules['confidence'] > 0.6) &
      (rules['lift'] > 5)
      ]
```
